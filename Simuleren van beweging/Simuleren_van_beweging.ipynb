```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def forward_euler():
    h = 0.1
    g = 9.81
    friction = 0.1

    num_steps=50
    t = np.zeros(num_steps + 1) # time
    x = np.zeros(num_steps + 1) # height
    v = np.zeros(num_steps + 1) # velocity

    for step in range(num_steps):
        t[step + 1] = t[step] + h
        x[step + 1] = x[step] + (h * v[step])
        v[step + 1] = (v[step] - (h * g)) * (1 - friction) # (g + k / m * v[step]**2)
    return t, x, v
```

```python
time, height, velocity = forward_euler()
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10,6))
axes[0].plot(time, height)
axes[0].set_ylabel("Height in m")
axes[0].set_xlabel("Time in s")
axes[1].plot(time, velocity)
axes[1].set_ylabel("Velocity in m/s2")
axes[1].set_xlabel("Time in s")
plt.show()
```

```python
def forward_euler_spring():
    h = 0.1
    spring_constant = 500
    mass = 50
    friction = 0.1

    num_steps = 1000
    t = np.zeros(num_steps + 1) # time
    x = np.zeros(num_steps + 1) # height
    v = np.zeros(num_steps + 1) # velocity

    x[0] = 10 # start
    for step in range(num_steps):
        force = -spring_constant * x[step] - v[step] 
        acceleration = force / mass
        t[step + 1] = t[step] + h
        x[step + 1] = x[step] + (h * v[step])
        v[step + 1] = (v[step] + (h * acceleration)) * (1 - friction)
        
    return t, x, v
```

```python
time, height, velocity = forward_euler_spring()
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10,6))
axes[0].plot(time, height)
axes[0].set_ylabel("Height in m")
axes[0].set_xlabel("Time in s")
axes[1].plot(time, velocity)
axes[1].set_ylabel("Velocity in m/s2")
axes[1].set_xlabel("Time in s")
plt.show()
```
