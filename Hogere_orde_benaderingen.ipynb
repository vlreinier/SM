```python
import math
import matplotlib.pylab as plt
import numpy as np
```

```python
def euler_heun(steps=100):
    euler = np.zeros([steps + 1])
    heun = np.zeros([steps + 1])
    eulers_number = np.zeros([steps + 1])

    h = 1 / steps
    t = list(np.arange(0, 1+h, step=h))

    euler[0] = 1
    heun[0] = 1
    eulers_number[0] = 1

    for step in range(steps):
        euler[step + 1] = euler[step] + h * euler[step]  # Euler
        heun[step + 1] = heun[step] + h * (euler[step] + euler[step + 1]) / 2  # Heun
        eulers_number[step + 1] = math.e ** t[step + 1]  # e^t
        
    return t, euler, heun, eulers_number
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(12,6))
for x in [5, 10, 20, 100]:
    t, euler, heun, eulers_number = euler_heun(steps=x)

    # Euler
    axes[0].title.set_text("Eulers method")
    axes[0].plot(t, euler, label="n = {}".format(x))
    axes[0].axhline(y=math.e, color='black', linestyle='--')
    axes[0].legend()

    # Heun
    axes[1].title.set_text("Heuns method")
    axes[1].plot(t, heun, label="n = {}".format(x))
    axes[1].axhline(y=math.e, color='black', linestyle='--')
    axes[1].legend()

# Eulers number calculated on 100 steps intervals
axes[2].title.set_text("Eulers number")
axes[2].plot(t, eulers_number, label="n = 100")
axes[2].axhline(y=math.e, color='black', linestyle='--')
axes[2].legend()

plt.tight_layout()
plt.show()
```
